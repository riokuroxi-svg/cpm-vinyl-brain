## 1. Sube `CPM_Vinyl_Brain_0.3_TestBuild.zip`

```python
from google.colab import files
import zipfile, pathlib, os
uploaded = files.upload()
package = next(name for name in uploaded if name.lower().endswith('.zip'))
with zipfile.ZipFile(package) as z: z.extractall('/content/cpmbrain')
project = pathlib.Path('/content/cpmbrain/CPM_Vinyl_Brain_0.3_TestBuild/project')
print('Proyecto:', project)
```

## 2. Sube una imagen PNG/JPG para probar

```python
uploaded_image = files.upload()
image_name = next(iter(uploaded_image))
image_path = pathlib.Path('/content') / image_name
image_path.write_bytes(uploaded_image[image_name])
print('Imagen:', image_path)
```

## 3. Ejecuta el Cerebro (perfil quality)

```python
import subprocess, sys, shutil
result = pathlib.Path('/content/resultado_cpm_brain')
if result.exists(): shutil.rmtree(result)
cmd = [sys.executable, str(project/'research/analyzer_v02/run_brain.py'),
       '--input', str(image_path), '--out', str(result),
       '--shapes', str(project/'src/main/resources/shapes'),
       '--profile', 'quality', '--seed', '20260720']
subprocess.run(cmd, check=True)
print('Terminado:', result)
```

## 4. Visualiza y descarga

```python
from IPython.display import display
from PIL import Image
display(Image.open(result/'04_optimize_reduce_v05/FINAL_COMPARISON_03_VS_05.png'))
archive = shutil.make_archive('/content/resultado_cpm_brain', 'zip', result)
files.download(archive)
```
